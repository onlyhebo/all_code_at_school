```python
url_list = []
f = open('zhihu.txt','w',encoding= 'utf8')
```

```python
for i in range(len(li)):
    print(i)
    url_list.append(li[i].find_element_by_css_selector('h2 a').get_attribute('href'))
    print(url_list[i])
```

```python
for i in range(len(url_list)):
    print(i)
    url = url_list[i]
    driver.get(url)
    print(url)
    a = True
    try:
        driver.find_element_by_css_selector('.QuestionMainAction')
        a = True
    except:
        a = False
    if a == True:
        more = driver.find_element_by_css_selector('.QuestionMainAction').text
        print(more)
        if more != '更多推荐结果':
            driver.find_element_by_css_selector('.QuestionMainAction').click()
            for i in range(10):
                driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
                time.sleep(2)
            content_list = driver.find_elements_by_css_selector('#QuestionAnswers-answers  .List .List-item')
            for j in range(len(content_list)):
                content = content_list[j].find_element_by_css_selector('.RichText.CopyrightRichText-richText').text.replace('\n','')
                f.write(content+'\n')
                print(content.replace('\n','') + '\n')
        else:
            print(more)
    else:
        print("没有更多，一个人比比")
```
