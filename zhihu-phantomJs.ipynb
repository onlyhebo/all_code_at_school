```python
for i in range(len(li)):
    url = li[i].find_element_by_css_selector('h2 a').get_attribute('href')
    driver.get(url)
    print(url)
    a = True
    try:
        driver.find_element_by_css_selector('.QuestionMainAction')
        a = True
    except:
        a = False
    if a == True:
        more = driver.find_element_by_css_selector('.QuestionMainAction').text
        print(more)
        if more != '更多推荐结果':
            driver.find_element_by_css_selector('.QuestionMainAction').click()
            content_list = driver.find_elements_by_css_selector('.List .List-item')
            for j in range(len(content_list)):
                print(content_list[j].find_element_by_css_selector('.RichText.CopyrightRichText-richText').text.replace('\n','') + '\n')
        else:
            print(more)
    else:
        print("没有更多，一个人比比")
```
